```python
import os
import warnings
warnings.simplefilter('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
#preprocessing data
```

```python
exoplanet = pd.read_csv('cumulative.csv')
exoplanet.head()
```

```python
exoplanet.count()
```

```python
exoplanet=exoplanet.drop(columns=["koi_fpflag_nt","koi_fpflag_ss","koi_fpflag_co","koi_teq_err1","koi_teq_err2"])
exoplanet.head()
```

```python
exoplanet=exoplanet.rename(columns={"kepid":"ID", "kepoi_name":"KOI Name",
                         "kepler_name":"Kepler Name","koi_disposition":"Exoplanet Archive",
                         "koi_pdisposition":"Kepler Disposition", "koi_score":"Score"})
exoplanet.head()
```

```python
exoplanet=exoplanet.drop(exoplanet.columns[0], axis=1)
exoplanet.head()
```

```python
exoplanet= exoplanet.dropna(axis='index')
exoplanet.head()
```

```python
X = exoplanet[["ID","KOI Name", "Exoplanet Archive", "Kepler Disposition"]]
y = exoplanet["Score"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.utils import to_categorical 
from keras.layers import Dense
from keras.datasets import mnist
```

```python
#splitting data
```

```python
data = X.copy()

data_binary_encoded = pd.get_dummies(data)
data_binary_encoded.head()
```

```python
#splitting into test and train data sets
```

```python
from sklearn.model_selection import train_test_split
X = pd.get_dummies(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train
```

```python
#model training 
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
 X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
plt.scatter(model.predict(X_train_scaled), model.predict(X_train_scaled) - y_train_scaled, c="blue", label="Training Data")
plt.scatter(model.predict(X_test_scaled), model.predict(X_test_scaled) - y_test_scaled, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test_scaled.min(), xmax=y_test_scaled.max())
plt.title("Residual Plot")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

predictions = model.predict(X_test_scaled)
MSE = mean_squared_error(y_test_scaled, predictions)
r2 = model.score(X_test_scaled, y_test_scaled)

print(f"MSE: {MSE}, R2: {r2}")
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
```

```python
 from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=100, centers=2, random_state=0, cluster_std=.95)
plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap="bwr");
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5 , 10],
              'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
predictions = grid.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["blue", "red"]))
```
